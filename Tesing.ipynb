```python
import pandas as pd
from fuzzywuzzy import fuzz
import random
import re

def generate_variations(plant_name):
    """Генерация различных вариаций названия растения"""
    variations = []
    
    # Оригинальное название
    variations.append(plant_name)
    
    # Только первое слово
    first_word = plant_name.split()[0]
    variations.append(first_word)
    
    # Сокращённая версия (первые 3-4 буквы первого слова)
    if len(first_word) > 3:
        variations.append(first_word[:3])
        variations.append(first_word[:4])
    
    # Типичные опечатки
    typos = plant_name.replace('о', 'а').replace('е', 'и')
    variations.append(typos)
    
    # Без буквы ё
    variations.append(plant_name.replace('ё', 'е'))
    
    # Добавление контекстных запросов
    context_queries = [
        f"где растет {plant_name}",
        f"покажи {plant_name}",
        f"что такое {plant_name}",
        f"описание {plant_name}",
        f"фото {plant_name}"
    ]
    variations.extend(context_queries)
    
    return variations

def create_test_cases(plant_list):
    """Создание структурированного набора тестовых случаев"""
    test_cases = {
        'exact_names': plant_list,
        'variations': [],
        'partial_names': [],
        'context_queries': [],
        'invalid_queries': [
            'Париж', 'слон', 'компьютер', 'машина', 'телефон',
            'где парк', 'как пройти', 'режим работы', 'билеты'
        ]
    }
    
    # Генерация вариаций для каждого растения
    for plant in plant_list:
        variations = generate_variations(plant)
        test_cases['variations'].extend(variations)
        
        # Создание частичных имен
        words = plant.split()
        if len(words) > 1:
            test_cases['partial_names'].append(words[0])
            if len(words[0]) > 3:
                test_cases['partial_names'].append(words[0][:3])
    
    return test_cases

def test_model_response(query, df_plants):
    """Тестирование ответа модели на запрос"""
    max_ratio = 0
    matched_name = None
    
    for name in df_plants['Name'].unique():
        ratio = fuzz.partial_ratio(name.lower(), query.lower())
        if ratio > max_ratio and ratio > 70:
            max_ratio = ratio
            matched_name = name
    
    return {
        'query': query,
        'matched_name': matched_name,
        'confidence': max_ratio,
        'success': matched_name is not None
    }

def run_comprehensive_tests(plant_list, df_plants):
    """Запуск комплексного тестирования"""
    test_cases = create_test_cases(plant_list)
    results = {category: [] for category in test_cases.keys()}
    
    print("Начало комплексного тестирования...")
    
    for category, queries in test_cases.items():
        print(f"\nТестирование категории: {category}")
        print("=" * 50)
        
        for query in queries:
            result = test_model_response(query, df_plants)
            results[category].append(result)
            
            print(f"\nЗапрос: {query}")
            print(f"Найдено: {result['matched_name']}")
            print(f"Уверенность: {result['confidence']}%")
    
    return results

def analyze_test_results(results):
    """Анализ результатов тестирования"""
    print("\nАнализ результатов:")
    print("=" * 50)
    
    for category, tests in results.items():
        successful = len([t for t in tests if t['success']])
        total = len(tests)
        
        print(f"\nКатегория: {category}")
        print(f"Всего тестов: {total}")
        print(f"Успешных совпадений: {successful}")
        print(f"Точность: {(successful/total)*100:.2f}%")
        
        # Анализ проблемных случаев
        if category != 'invalid_queries':
            failed = [t for t in tests if not t['success']]
            if failed:
                print("\nПроблемные запросы:")
                for f in failed[:5]:  # Показываем первые 5 проблемных случаев
                    print(f"Запрос: {f['query']}")

if __name__ == '__main__':
    # Загрузка данных
    CSV_PATH = r"C:\Users\Mikl\zaryadye_bot\60861CSV\data-60861-2024-08-06.csv"
    df_plants = pd.read_csv(CSV_PATH, sep=';', encoding='utf-8', on_bad_lines='skip')
    df_plants = df_plants[df_plants['ID'] != 'Код'].reset_index(drop=True)
    
    # Получение списка растений
    plant_list = df_plants['Name'].unique().tolist()
    
    # Запуск тестирования
    results = run_comprehensive_tests(plant_list, df_plants)
    
    # Анализ результатов
    analyze_test_results(results)
```

Этот код:

Выводит только сводную статистику
Защищен от деления на ноль
Включает тесты нерелевантных слов
Показывает неадекватные ответы модели
Предоставляет метрики по категориям
Вы получите краткий отчет с:

Общей точностью модели
Метриками по категориям
Списком проблемных ответов
Анализом реакции на нерелевантные запросы

```python
import pandas as pd
from fuzzywuzzy import fuzz
from collections import defaultdict

def create_test_cases(plant_list):
    """Создание структурированного набора тестовых случаев"""
    non_plant_words = [
        'париж', 'слон', 'самолет', 'зона', 'климат', 
        'цветение', 'шоколад', 'европа', 'компьютер'
    ]
    
    test_cases = {
        'exact_names': plant_list,
        'partial_names': [name.split()[0] for name in plant_list],
        'misspelled': [name.replace('о', 'а') for name in plant_list[:10]],
        'non_plant_queries': non_plant_words,
        'context_queries': [
            f"где растет {plant}" for plant in plant_list[:5]
        ]
    }
    return test_cases

def test_model_response(query, df_plants):
    max_ratio = 0
    matched_name = None
    
    for name in df_plants['Name'].unique():
        ratio = fuzz.partial_ratio(name.lower(), query.lower())
        if ratio > max_ratio and ratio > 70:
            max_ratio = ratio
            matched_name = name
    
    return {
        'query': query,
        'matched_name': matched_name,
        'confidence': max_ratio,
        'success': matched_name is not None
    }

def run_final_tests(df_plants):
    plant_list = df_plants['Name'].unique().tolist()
    test_cases = create_test_cases(plant_list)
    
    results = defaultdict(list)
    metrics = defaultdict(dict)
    unusual_responses = []
    
    # Выполнение тестов
    for category, queries in test_cases.items():
        if queries:  # Проверка на пустой список
            successful = 0
            total = len(queries)
            
            for query in queries:
                result = test_model_response(query, df_plants)
                results[category].append(result)
                
                if result['success']:
                    successful += 1
                    
                    # Проверка неадекватных ответов
                    if category == 'non_plant_queries' and result['confidence'] > 70:
                        unusual_responses.append({
                            'query': query,
                            'matched': result['matched_name'],
                            'confidence': result['confidence']
                        })
            
            if total > 0:  # Защита от деления на ноль
                metrics[category] = {
                    'total': total,
                    'successful': successful,
                    'accuracy': (successful/total)*100
                }
    
    return metrics, unusual_responses

def print_final_report(metrics, unusual_responses):
    print("\nСВОДНЫЙ ОТЧЕТ ТЕСТИРОВАНИЯ")
    print("=" * 50)
    
    # Общая метрика
    total_tests = sum(m['total'] for m in metrics.values())
    total_successful = sum(m['successful'] for m in metrics.values())
    overall_accuracy = (total_successful/total_tests)*100 if total_tests > 0 else 0
    
    print(f"\nОБЩАЯ ТОЧНОСТЬ: {overall_accuracy:.2f}%")
    print(f"Всего тестов: {total_tests}")
    print(f"Успешных совпадений: {total_successful}")
    
    print("\nМЕТРИКИ ПО КАТЕГОРИЯМ:")
    for category, metric in metrics.items():
        print(f"\n{category}:")
        print(f"Точность: {metric['accuracy']:.2f}%")
        print(f"Тестов: {metric['total']}")
        print(f"Успешных: {metric['successful']}")
    
    if unusual_responses:
        print("\nНЕАДЕКВАТНЫЕ ОТВЕТЫ НА НЕРЕЛЕВАНТНЫЕ ЗАПРОСЫ:")
        print("-" * 50)
        for resp in unusual_responses:
            print(f"Запрос: {resp['query']}")
            print(f"Ошибочное совпадение: {resp['matched']}")
            print(f"Уверенность: {resp['confidence']}%\n")

if __name__ == '__main__':
    # Загрузка данных
    CSV_PATH = r"C:\Users\Mikl\zaryadye_bot\60861CSV\data-60861-2024-08-06.csv"
    df_plants = pd.read_csv(CSV_PATH, sep=';', encoding='utf-8', on_bad_lines='skip')
    df_plants = df_plants[df_plants['ID'] != 'Код'].reset_index(drop=True)
    
    # Запуск тестирования
    metrics, unusual_responses = run_final_tests(df_plants)
    
    # Вывод отчета
    print_final_report(metrics, unusual_responses)

```
